```python
panjang_balok = float(input("Masukan nilai panjang dari balok (cm): "))
lebar_balok = int(input("Masukan nilai lebar dari balok (cm): "))
tinggi_balok = float(input("Masukan nilai tinggi dari balok (cm): "))

volume_balok = panjang_balok * lebar_balok * tinggi_balok

print(f"Volume balok yaitu: {volume_balok} cm^3")
```

##Aplikasi Sederhana Menghitung Volume Bola

```python
jari_jari = int(input("Masukkan jari-jari dari bola (cm): "))

pi = 22/7

volume_bola = 4/3 * pi * jari_jari**3

print(f"Volume bola adalah {volume_bola} cm^3. ")
```

#Aplikasi Menghitung Volume Tabung

```python
jari_jari = int(input("Masukkan jari-jari dari tabung (cm): "))
pi = 22/7
tinggi = int(input("Masukkan tinggi dari tabung (cm): "))
volume_tabung = pi * jari_jari**2 * tinggi
print(f"Volume tabung adalah {volume_tabung} cm^3. ")
```
